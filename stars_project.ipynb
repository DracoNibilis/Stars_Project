### A theoretical background.

Stars are the most widely recognized astronomical object, and represent the most fundamental building blocks of galaxies. 
Stars are born within the clouds of dust. By turbulence deep within these clouds knots are created with sufficient mass that the gas and dust can begin to collapce under its own gravitational attraction.
As a star collapse, th ematerial at the center begins to heat up, becoming protostar. And that hot core will become star.

Stars are long lifed and dynamic objects, on human time scale it is hard to observe those changes. All the stars thet we see seems constant, and on the observation that we made we can classified them. 

In astronomy , stars classification is based on their spectral characteristics. 



#### Types of classification

- Hertzsprung-Russel -

It is a diagram, a scatter plot of stars showing the relationship between the stars absolute magnitudes or luminosities versus their stellar classifications of effective temperatures.



#### Harvard Classification
The Harvard system is a one-dimensional classification scheme by astronomer Annie Jump Cannon. Stars are grouped according to their spectral characteristics by single letters of the alphabet, optionally with numeric subdivisions.

classes:
- O - O-type stars are very hot and extremely luminous, with most of their radiated output in the ultraviolet range. These are the rarest of all main-sequence stars. About 1 in 3,000,000 (***0.00003%***) of the main-sequence stars in the solar neighborhood are O-type stars.[e][8] Some of the most massive stars lie within this spectral class. O-type stars frequently have complicated surroundings that make measurement of their spectra difficult. 
- B - B-type stars are very luminous and blue. Their spectra have neutral helium lines, which are most prominent at the B2 subclass, and moderate hydrogen lines. As O- and B-type stars are so energetic, they only live for a relatively short time. Thus, due to the low probability of kinematic interaction during their lifetime, they are unable to stray far from the area in which they formed, apart from runaway stars. About 1 in 800 (***0.125%***) of the main-sequence stars in the solar neighborhood are B-type main-sequence stars
- A - A-type stars are among the more common naked eye stars, and are white or bluish-white. They have strong hydrogen lines, at a maximum by A0, and also lines of ionized metals (Fe II, Mg II, Si II) at a maximum at A5. The presence of Ca II lines is notably strengthening by this point. About 1 in 160 (***0.625%***) of the main-sequence stars in the solar neighborhood are A-type stars.
- F - F-type stars have strengthening spectral lines H and K of Ca II. Neutral metals (Fe I, Cr I) beginning to gain on ionized metal lines by late F. Their spectra are characterized by the weaker hydrogen lines and ionized metals. Their color is white. About 1 in 33 (***3.03%***) of the main-sequence stars in the solar neighborhood are F-type stars.
- G - G-type stars, including the Sun,[10] have prominent spectral lines H and K of Ca II, which are most pronounced at G2. They have even weaker hydrogen lines than F, but along with the ionized metals, they have neutral metals. There is a prominent spike in the G band of CN molecules. Class G main-sequence stars make up about ***7.5%***, nearly 1 -13, of the main-sequence stars in the solar neighborhood.
- K - K-type stars are orangish stars that are slightly cooler than the Sun. They make up about ***12%*** of the main-sequence stars in the solar neighborhood.[e][8] There are also giant K-type stars, which range from hypergiants like RW Cephei, to giants and supergiants, such as Arcturus, whereas orange dwarfs, like Alpha Centauri B, are main-sequence stars. 
- M - Class M stars are by far the most common. About ***76%*** of the main-sequence stars in the solar neighborhood are class M stars.[e][f][8] However, class M main-sequence stars (red dwarfs) have such low luminosities that none are bright enough to be seen with the unaided eye, unless under exceptional conditions. The brightest-known M class main-sequence star is Lacaille 8760, class M0V, with magnitude 6.7 (the limiting magnitude for typical naked-eye visibility under good conditions is typically quoted as 6.5), and it is extremely unlikely that any brighter examples will be found.

### Creating variables.

I will try to create variables 
Lets create variables, and way to do that, I will strat from importing usefull modules.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Create list of columns.
columns = ['Temperature', 'Mass', 'Radius', 'Luminosity', 'Class']

# Create empty data set.
df = pd.DataFrame(columns = columns)

# Check data set.
df.head()
````

#### Temperature.

I will start from creating variable `Temperature`, since it is most important variable, and it will directly determinated the 'letter variable' from Morgan-Keenan classification.

Temperature variable in original data set is seperate in 7 groups:<br>

```
- >= 30000
- 10000 - 30000
- 7500 - 10000
- 6000 - 7500
- 5200 - 6000
- 3700 - 5200
- 2400 - 3700
```

Based on those groups I wil ltry to create my own data.

Lets create set of ranges that will be used for generating values.

````python
# Generate values for each range of temperature variable.
# Setting seed to have the same set with each generation.
np.random.seed(1212)

# Generating 500 data points.
# NOTE: Size of groups was choosen aproximetly to the true set of classification.
a = list(np.random.randint(2400, 3700, 300))
b = list(np.random.randint(3700, 5200, 100))
c = list(np.random.randint(5200, 6000, 40))
d = list(np.random.randint(6000, 7500, 30))
e = list(np.random.randint(7500, 10000, 20))
f = list(np.random.randint(10000, 30000, 5))
g = list(np.random.randint(30000, 100000, 5))

# Creating list of generated values.
list_of_temp = [a,b,c,d,e,f,g]

# Creating empty list for variable temperature.
temperature = list()

# Check the value in list_of_temp.
# Checking just few values, not whole set.
print(list_of_temp[:1])
````

````python
# Fill up temperature list with generated data.
for i in list_of_temp:
    for j in i:
        temperature.append(j)
        
# Check if the list is generating without errors.
print("TEMPERATURE LIST TEST ", temperature[:30])

# Check the length of the temperature list.
print(len(temperature))
````

````python
# Sorting values in the temperature list in descending order.
temperature.sort(reverse=True)

# Check temperature list if sort worked.
print("First 20 of records in temperature set are: ", temperature[:20])
````

So now I can say that my first variable is ready. We can add it to data set and start prepering others variariable.

````python
# Add first variable to data set.
df['Temperature'] = temperature

# Check data set if data were added.
df.head()
````

***NOTE:*** My stars collection will be loosely connected with original Milky Way data. Only in first variable I will try to stick more with original data set.

````python
# Add mass variable to data set.
df['Mass'] = mass

# Check data set if mass data were added.
df.head()
````

#### Luminosity.

Luminosity variable in the original stars classification set is also close into few ranges. Values are range from 0.08 to 30000, and ranges are spread as follows:
```
    >= 30000
    25 - 30000
    5 - 25
```

Also, here I have to mention that ranges till 1.5 have higher probability to accur than those above 1.5. As in previous variables I'm gonna use `np.random.exponential` ans set the parameter for 0.6, we will see what kind of variables will be generated.

````python
# Genereting variable "luminosity".
luminosity = np.random.exponential(0.6, 500)

# Check if data were generated, just sample from full set.
print(luminosity[:20])
````

````python
# Add luminosity variable to data set.
df['Luminosity'] = luminosity

# Check data set if luminosity data were added.
df.head()
````

As I added all numerical variables to the data set, now I will add the ***class*** variable, which is based on the temperature variable.

#### Class.

***NOTE***<br>
Class values are strictly connected with temperature values. So here I will fill up ***Class*** column with data assigned by conditions. Just as a reminder lets look again at the groups that temperature data are put in:<br>
```
- >= 30000
- 10000 - 30000
- 7500 - 10000
- 6000 - 7500
- 5200 - 6000
- 3700 - 5200
- 2400 - 3700
```

````python
# Create a list of conditions. 
conditions = [
    (df['Temperature'] > 2400) & (df['Temperature'] < 3700),
    (df['Temperature'] > 3700) & (df['Temperature'] < 5200),
    (df['Temperature'] > 5200) & (df['Temperature'] < 6000),
    (df['Temperature'] > 6000) & (df['Temperature'] < 7500),
    (df['Temperature'] > 7500) & (df['Temperature'] < 10000),
    (df['Temperature'] > 10000) & (df['Temperature'] < 30000),
    (df['Temperature'] > 30000)
             ]

# Create class data list that we want to assign to conditions.
classes = ["M", "K", "G", "F", "A", "B", "O"]

# Add to column 'Class' choosen data by condition.
df['Class'] = np.select(conditions, classes)

# Check if data are correctly imported to dataset.
df.head(10)
````

````python
# Plotting more data.
fig = plt.figure(figsize=(6, 6))
plt.scatter(df['Mass'], df['Luminosity'],
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 50,
           c=df['Temperature'])

# Describe labels.
plt.xlabel("Mass")
plt.ylabel("Luminosity")

plt.colorbar()
plt.title("Stars Mass,Luminosity, Temperature Scatter Plot ")
plt.show()
````
